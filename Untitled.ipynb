```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics

%matplotlib inline
```

```python
df = pd.DataFrame()
df = pd.read_csv('football.csv')
```

```python
df.head()
```

```python
df.fillna(0)
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df.head()
```

```python
mean_wage_by_finishing = df.groupby(by='Finishing')['Wage'].mean()
```

```python
#Чем выше уровень доводки мяча, тем выше зарплата

# инициализируем фигуру и указываем ее размер
plt.figure(figsize=(10, 6))

# название графика
plt.title('Изменение средней зарплаты по доводке', fontsize=16)

# строим график
plt.plot(mean_wage_by_finishing, color='green', linestyle='-.', label='Зарплата')

# настраиваем подписи осей
plt.ylabel('Средняя зарплата', fontsize=16)
plt.xlabel('Доводка', fontsize=16)

# включаем легенду и сетку
plt.legend(fontsize=14)
plt.grid()
```

```python
#Самая популярная позиция - вратарь

df['Position'].value_counts().plot(kind='bar', )
plt.title('Распределение по позициям');
```

```python
mean_age_by_positioning = df.groupby(by='Age')['Positioning'].mean()
```

```python
# Уровень позиционирования растёт с возрастом, но по достижению определённой точки прекращает расти и падает

# инициализируем фигуру и указываем ее размер
plt.figure(figsize=(10, 6))

# название графика
plt.title('Изменение среднего позиционирования по возрасту', fontsize=16)

# строим график
plt.plot(mean_age_by_positioning, color='green', linestyle='-.', label='Позиционирование')

# настраиваем подписи осей
plt.ylabel('Среднее позиционирование', fontsize=16)
plt.xlabel('Возраст', fontsize=16)

# включаем легенду и сетку
plt.legend(fontsize=14)
plt.grid()
```
